```python
minmax_scale = MinMaxScaler(feature_range=(0, 1))
def normalization(arquivo,col):
  for i in col:
    arr = arquivo[i]
    arr = np.array(arr)
    arquivo[i] = minmax_scale.fit_transform(arr.reshape(len(arr),1))
  return arquivo
```

```python
matriz_correlacao = arquivo.corr()

```

```python
# Define o limite de correlação
limite_correlacao = 0.95

# Cria uma lista para armazenar as colunas a serem removidas
colunas_remover = []

# Percorre a matriz de correlação
for i in range(len(matriz_correlacao.columns)):
    for j in range(i):
        # Verifica se a correlação é maior do que o limite
        if abs(matriz_correlacao.iloc[i, j]) > limite_correlacao:
            coluna1 = matriz_correlacao.columns[i]
            coluna2 = matriz_correlacao.columns[j]
            print(f'As colunas "{coluna1}" e "{coluna2}" estão altamente correlacionadas ({matriz_correlacao.iloc[i, j]:.2f}).')
            # Adiciona uma das colunas à lista de colunas a serem removidas
            if coluna1 not in colunas_remover:
                colunas_remover.append(coluna1)

# Remove as colunas altamente correlacionadas do DataFrame
arquivo = arquivo.drop(colunas_remover, axis=1)
print(f'\nForam removidas {len(colunas_remover)} colunas altamente correlacionadas.')
print(f'O novo DataFrame tem {arquivo.shape[1]} colunas.')
```

```python
# calcular as importâncias das características para cada classe
importance = modelo.feature_importances_
nomes_caracteristicas = X.columns
importance_por_classe = {}
for classe in np.unique(y_treino):
    mask = (y_treino == classe)
    importance_por_classe[classe] = np.mean(importance[mask], axis=0)

# classificar as características de acordo com a sua importância média para cada classe
caracteristicas_por_classe = {}
for classe in importance_por_classe:
    idxs = np.argsort(importance_por_classe[classe])[::-1]
    caracteristicas_por_classe[classe] = nomes_caracteristicas[idxs]

# selecionar as principais características para cada classe com base na sua importância média
n_caracteristicas = 2
principais_caracteristicas_por_classe = {}
for classe in caracteristicas_por_classe:
    principais_caracteristicas_por_classe[classe] = caracteristicas_por_classe[classe][:n_caracteristicas]

print(principais_caracteristicas_por_classe)
```
